```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
base_url = 'https://understat.com/league/EPL/2020'
```

```python
#Use requests to get the webpage and BeautifulSoup to parse the page ##

## I have already scraped 2006 to 2020 premier league data from website##
##Here is an example of scraping ## 
##I will then append this dataset to my Originally scraped datset ##

## I wanted to test myself by exploring with the data and finding differnt ways to clean and transform it. ##

res = requests.get(base_url)
soup = BeautifulSoup(res.content, 'lxml')
divs = soup.find_all("div")
```

```python
scripts = soup.find_all('script')
```

```python
strings = scripts[1].string
```

```python
#Strip Symbols so I only have Json data ##
ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
data = json.loads(json_data)
```

```python
pd.Series(data[0],)
```

```python
df = pd.DataFrame(columns=["id", "isResult" ,"h", "a", 'goals', 'xG', 'datetime', 'forecast'])
```

```python
##Creating the Data Frame using a for loop
for i in range(0,len(data)) :
    currentItem = data[i]
    df.loc[i] = [data[i]["id"], data[i]["isResult"],data[i]["h"],data[i]["a"],data[i]["goals"],data[i]["xG"],data[i]["datetime"],data[i]["forecast"]]
```

```python
df
```

```python
columnsplit=df['h'].apply(pd.Series)
```

```python
columnsplit['title']
```

```python
df['Home_Club'] = columnsplit['title']
```

```python
columnsplitaway=df['a'].apply(pd.Series)
columnsplitaway['title']
df['Away_Club'] = columnsplitaway['title']
```

```python
new_df = df[['Home_Club', 'Away_Club', 'goals', 'xG', 'datetime', 'forecast']]
```

```python
new_df
```

```python
splitgoals = new_df['goals'].apply(pd.Series)
```

```python
home_goals= splitgoals['h']
away_goals= splitgoals['a']
new_df['home_goals'] = home_goals
new_df['away_goals'] = away_goals
```

```python
splitxG = new_df['xG'].apply(pd.Series)
```

```python
xHomeG= splitxG['h']
xAwayG= splitxG['a']
new_df['xHomeG'] = xHomeG
new_df['xAwayG'] = xAwayG
```

```python
new_df
```

```python
splitForecast = new_df['forecast'].apply(pd.Series)
```

```python
winforecast= splitForecast['w']
loseforecast= splitForecast['l']
drawforecast= splitForecast['d']

new_df['winforecast'] = winforecast
new_df['loseforecast'] = loseforecast
new_df['drawforecast'] = drawforecast
```

```python
new_df
```

```python
df = new_df[['Home_Club', 'Away_Club', 'home_goals', 'away_goals','datetime' ,'xHomeG', 'xAwayG', 'winforecast', 'loseforecast', 'drawforecast']]
```

```python
df.head
```

```python
##Creating the record
df.loc[df['home_goals'] > df['away_goals'], 'RESULT'] = 'H'
df.loc[df['home_goals'] == df['away_goals'], 'RESULT'] = 'D'
df.loc[df['home_goals'] < df['away_goals'], 'RESULT'] = 'A'
```
